```python
# STEP 1: IMPORT LIBRARIES

import pandas as pd
import numpy as np
```

```python
# STEP 2: CREATE REALISTIC HEALTHCARE DATASET

np.random.seed(42)

vendors = ['Vendor_A', 'Vendor_B', 'Vendor_C']
products = ['Gloves', 'Syringes', 'IV Kits', 'Masks', 'Bandages']
hospitals = ['Hospital_X', 'Hospital_Y', 'Hospital_Z']

data = []

for _ in range(1000):
    product = np.random.choice(products)
    vendor = np.random.choice(vendors)
    hospital = np.random.choice(hospitals)

    base_price = {
        'Gloves': 5,
        'Syringes': 10,
        'IV Kits': 50,
        'Masks': 2,
        'Bandages': 3
    }[product]

    price = round(base_price * np.random.uniform(0.8, 1.5), 2)
    contract_price = round(base_price * np.random.uniform(0.7, 1.0), 2)

    data.append([hospital, vendor, product, price, contract_price])

df = pd.DataFrame(data, columns=['Hospital', 'Vendor', 'Product', 'Price', 'Contract_Price'])
```

```python
# STEP 3: IDENTIFY PRICE VARIANCE

df['Price_Variance'] = df['Price'] - df['Contract_Price']
```

```python
# STEP 4: IDENTIFY SAVINGS OPPORTUNITIES

df['Savings_Opportunity'] = df['Price_Variance'].apply(lambda x: x if x > 0 else 0)
```

```python
# STEP 5: CONTRACT COMPLIANCE

df['Contract_Compliance'] = df['Price'] <= df['Contract_Price']
```

```python
# STEP 6: AGGREGATE INSIGHTS

summary = df.groupby('Product').agg({
    'Savings_Opportunity': 'sum',
    'Price': 'mean',
    'Contract_Price': 'mean'
}).reset_index()

print("=== Savings Summary by Product ===")
print(summary)
```

```python
# STEP 7: TOTAL SAVINGS

total_savings = df['Savings_Opportunity'].sum()
print(f"\nTotal Potential Savings Identified: ${total_savings:.2f}")
```

```python
# STEP 8: EXPORT DATA FOR POWER BI

df.to_csv('pricing_analysis.csv', index=False)
summary.to_csv('pricing_summary.csv', index=False)

print("\nCSV files exported for dashboarding.")
```
